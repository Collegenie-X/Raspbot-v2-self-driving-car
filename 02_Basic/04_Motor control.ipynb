```python
## Raspbot ë“œë¼ì´ë¸Œ ë¼ì´ë¸ŒëŸ¬ë¦¬ ë° ê¸°íƒ€ ê´€ë ¨ ë¼ì´ë¸ŒëŸ¬ë¦¬ import
import time
from Raspbot_Lib import Raspbot
from ipywidgets import interact
import ipywidgets as widgets
```

```python
# Raspbot ê°ì²´ bot ìƒì„±
bot = Raspbot()
```

```python
# ëª¨í„° ë™ì‘ ì œì–´
def run_motor(M1, M2, M3, M4):
    bot.Ctrl_Muto(0, M1)
    bot.Ctrl_Muto(1, M2)
    bot.Ctrl_Muto(2, M3)
    bot.Ctrl_Muto(3, M4)
    return True


# 4ê°œì˜ ìŠ¬ë¼ì´ë”ë¥¼ ìƒì„±í•˜ì—¬ ëª¨í„° ì œì–´
interact(
    run_motor,
    M1=widgets.IntSlider(min=-255, max=255, step=1, value=0),
    M2=widgets.IntSlider(min=-255, max=255, step=1, value=0),
    M3=widgets.IntSlider(min=-255, max=255, step=1, value=0),
    M4=widgets.IntSlider(min=-255, max=255, step=1, value=0),
)
```

```python
# ============================================
# ì°¨ëŸ‰ ì œì–´ í•¨ìˆ˜
# ============================================


def car_run(speed_left, speed_right):
    """
    ì§ì§„

    ë§¤ê°œë³€ìˆ˜:
    - speed_left: ì™¼ìª½ ëª¨í„° ì†ë„ (-255 ~ 255)
    - speed_right: ì˜¤ë¥¸ìª½ ëª¨í„° ì†ë„ (-255 ~ 255)
    """
    bot.Ctrl_Muto(0, speed_left)  # M1 (ì™¼ìª½ ì•)
    bot.Ctrl_Muto(1, speed_left)  # M2 (ì™¼ìª½ ë’¤)
    bot.Ctrl_Muto(2, speed_right)  # M3 (ì˜¤ë¥¸ìª½ ì•)
    bot.Ctrl_Muto(3, speed_right)  # M4 (ì˜¤ë¥¸ìª½ ë’¤)
    print(f"Forward: Left={speed_left}, Right={speed_right}")


def car_back(speed_left, speed_right):
    """
    í›„ì§„

    ë§¤ê°œë³€ìˆ˜:
    - speed_left: ì™¼ìª½ ëª¨í„° ì†ë„ (-255 ~ 255, ìŒìˆ˜ ì‚¬ìš©)
    - speed_right: ì˜¤ë¥¸ìª½ ëª¨í„° ì†ë„ (-255 ~ 255, ìŒìˆ˜ ì‚¬ìš©)
    """
    bot.Ctrl_Muto(0, -speed_left)  # M1 (ì™¼ìª½ ì•) - ìŒìˆ˜ë¡œ í›„ì§„
    bot.Ctrl_Muto(1, -speed_left)  # M2 (ì™¼ìª½ ë’¤) - ìŒìˆ˜ë¡œ í›„ì§„
    bot.Ctrl_Muto(2, -speed_right)  # M3 (ì˜¤ë¥¸ìª½ ì•) - ìŒìˆ˜ë¡œ í›„ì§„
    bot.Ctrl_Muto(3, -speed_right)  # M4 (ì˜¤ë¥¸ìª½ ë’¤) - ìŒìˆ˜ë¡œ í›„ì§„
    print(f"Backward: Left={-speed_left}, Right={-speed_right}")


def car_left(speed_left, speed_right):
    """
    ì¢ŒíšŒì „
    ì™¼ìª½ ëª¨í„°ëŠ” í›„ì§„, ì˜¤ë¥¸ìª½ ëª¨í„°ëŠ” ì „ì§„

    ë§¤ê°œë³€ìˆ˜:
    - speed_left: ì™¼ìª½ ëª¨í„° ì†ë„ (0 ~ 255, ìŒìˆ˜ë¡œ ë³€í™˜ë¨)
    - speed_right: ì˜¤ë¥¸ìª½ ëª¨í„° ì†ë„ (0 ~ 255)
    """
    bot.Ctrl_Muto(0, -speed_left)  # M1 (ì™¼ìª½ ì•) - í›„ì§„
    bot.Ctrl_Muto(1, -speed_left)  # M2 (ì™¼ìª½ ë’¤) - í›„ì§„
    bot.Ctrl_Muto(2, speed_right)  # M3 (ì˜¤ë¥¸ìª½ ì•) - ì „ì§„
    bot.Ctrl_Muto(3, speed_right)  # M4 (ì˜¤ë¥¸ìª½ ë’¤) - ì „ì§„
    print(f"Turn Left: Left={-speed_left}(backward), Right={speed_right}(forward)")


def car_right(speed_left, speed_right):
    """
    ìš°íšŒì „
    ì™¼ìª½ ëª¨í„°ëŠ” ì „ì§„, ì˜¤ë¥¸ìª½ ëª¨í„°ëŠ” í›„ì§„

    ë§¤ê°œë³€ìˆ˜:
    - speed_left: ì™¼ìª½ ëª¨í„° ì†ë„ (0 ~ 255)
    - speed_right: ì˜¤ë¥¸ìª½ ëª¨í„° ì†ë„ (0 ~ 255, ìŒìˆ˜ë¡œ ë³€í™˜ë¨)
    """
    bot.Ctrl_Muto(0, speed_left)  # M1 (ì™¼ìª½ ì•) - ì „ì§„
    bot.Ctrl_Muto(1, speed_left)  # M2 (ì™¼ìª½ ë’¤) - ì „ì§„
    bot.Ctrl_Muto(2, -speed_right)  # M3 (ì˜¤ë¥¸ìª½ ì•) - í›„ì§„
    bot.Ctrl_Muto(3, -speed_right)  # M4 (ì˜¤ë¥¸ìª½ ë’¤) - í›„ì§„
    print(f"Turn Right: Left={speed_left}(forward), Right={-speed_right}(backward)")


def car_stop():
    """
    ì •ì§€
    ëª¨ë“  ëª¨í„°ë¥¼ 0ìœ¼ë¡œ ì„¤ì •
    """
    for i in range(4):
        bot.Ctrl_Muto(i, 0)
    print("Stop")


# ============================================
# ì‚¬ìš© ì˜ˆì œ
# ============================================

print("âœ… Vehicle control functions are ready!")
print("\nUsage examples:")
print("  car_run(100, 100)      # Forward (Left=100, Right=100)")
print("  car_back(80, 80)       # Backward (Left=-80, Right=-80)")
print("  car_left(50, 100)      # Turn Left (Left=-50, Right=100)")
print("  car_right(100, 50)     # Turn Right (Left=100, Right=-50)")
print("  car_stop()             # Stop")
```
