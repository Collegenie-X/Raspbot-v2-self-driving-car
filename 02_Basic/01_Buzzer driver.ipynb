```python
# Import the Raspbot library
import time
from Raspbot_Lib import Raspbot
```

```python
# Create the Raspbot object bot
bot = Raspbot()
```

```python
# ============================================
# Buzzer Control Functions
# ============================================


def buzzer_on():
    """
    Turn on the buzzer
    """
    bot.Ctrl_BEEP_Switch(1)
    print("Buzzer: ON")


def buzzer_off():
    """
    Turn off the buzzer
    """
    bot.Ctrl_BEEP_Switch(0)
    print("Buzzer: OFF")


def buzzer_beep(duration=0.5):
    """
    Beep the buzzer for a specified duration

    Parameters:
    - duration: Beep duration in seconds (default: 0.5)
    """
    bot.Ctrl_BEEP_Switch(1)
    time.sleep(duration)
    bot.Ctrl_BEEP_Switch(0)
    print(f"Buzzer: Beeped for {duration} seconds")


# ============================================
# Usage Examples
# ============================================

print("âœ… Buzzer control functions are ready!")
print("\nUsage examples:")
print("  buzzer_on()              # Turn on buzzer")
print("  buzzer_off()              # Turn off buzzer")
print("  buzzer_beep(0.5)         # Beep for 0.5 seconds")
print("  buzzer_beep(1.0)         # Beep for 1.0 seconds")
```
